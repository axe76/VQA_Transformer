```python
import collections
import operator
# read the json file
annotation_file = 'v2_mscoco_train2014_annotations.json'
with open(annotation_file, 'r') as f:
    annotations = json.load(f)

# storing the captions and the image name in vectors
all_answers = []
all_answers_qids = []
all_img_name_vector = []
all_accepted_answers = []
for annot in annotations['annotations']:
#     print(annot)
#     break
    ans_dic = collections.defaultdict(int)
    for each in annot['answers']:
      diffans = each['answer']
      if diffans in ans_dic:
        #print(each['answer_confidence'])
        if each['answer_confidence']=='yes':
          ans_dic[diffans]+=4
        if each['answer_confidence']=='maybe':
          ans_dic[diffans]+= 2
        if each['answer_confidence']=='no':
          ans_dic[diffans]+= 1
      else:
        if each['answer_confidence']=='yes':
          ans_dic[diffans]= 4
        if each['answer_confidence']=='maybe':
          ans_dic[diffans]= 2
        if each['answer_confidence']=='no':
          ans_dic[diffans]= 1
#     print(ans_dic.items()) 
#     break
    all_accepted_answers.append(['~ '+a[0] +' `' for a in ans_dic.items()])
    most_fav = max(ans_dic.items(), key=operator.itemgetter(1))[0]
    #print(most_fav)
    caption = '~ ' + most_fav + ' `' #each['answer']
    
    image_id = annot['image_id']
    question_id = annot['question_id']
    full_coco_image_path = PATH + 'COCO_train2014_' + '%012d.jpg' % (image_id)

    all_img_name_vector.append(full_coco_image_path)
    all_answers.append(caption)
    all_answers_qids.append(question_id)
```

```python
def load_image(image_path):
    img = tf.io.read_file(image_path)
    img = tf.image.decode_jpeg(img, channels=3)
    #224 x 224 for VGG 299x299 for Inception
    img = tf.image.resize(img, (224, 224))
    img = tf.keras.applications.inception_v3.preprocess_input(img)
    return img, image_path
```

```python
def gru(units):
  # If you have a GPU, we recommend using CuDNNGRU(provides a 3x speedup than GRU)
  # the code automatically does that.
  return tf.keras.layers.GRU(units, 
                               return_sequences=True, 
                               return_state=True, 
                               recurrent_activation='sigmoid', 
                               recurrent_initializer='glorot_uniform')
```

```python
class Encoder(tf.keras.Model):
    def __init__(self):
        super(Encoder, self).__init__()
        self.bert_preprocess = hub.KerasLayer(tfhub_handle_preprocess)
        self.bert_model = hub.KerasLayer(tfhub_handle_encoder)
        
    def call(self, x, hidden):#hidden shape = (b,49,512)
        x = self.bert_preprocess(x)
        x = self.bert_model(x)["pooled_output"]
        return hidden, x
```

```python
class Decoder(tf.keras.Model):
    def __init__(self, vocab_size=30522, dec_units=512):
        super(Decoder, self).__init__()
        self.dec_units = dec_units
        self.bert_preprocess = hub.KerasLayer(tfhub_handle_preprocess)
        self.bert_model = hub.KerasLayer(tfhub_handle_encoder)
        self.gru = gru(self.dec_units)
        self.fc = tf.keras.layers.Dense(vocab_size)
        
        # used for attention
        self.W1 = tf.keras.layers.Dense(self.dec_units)
        self.W2 = tf.keras.layers.Dense(self.dec_units)
        self.V = tf.keras.layers.Dense(1)
        
    def call(self, x, hidden, enc_output):
        # enc_output shape == (batch_size, max_length, hidden_size)
        
        # hidden shape == (batch_size, hidden size)
        # hidden_with_time_axis shape == (batch_size, 1, hidden size)
        # we are doing this to perform addition to calculate the score
        print("Here 1")
        hidden_with_time_axis = tf.expand_dims(hidden, 1)
        
        print("Here 2")
        # score shape == (batch_size, max_length, hidden_size)
        score = tf.nn.tanh(self.W1(enc_output) + self.W2(hidden_with_time_axis))
        
        print("Here 3")
        # attention_weights shape == (batch_size, max_length, 1)
        # we get 1 at the last axis because we are applying score to self.V
        attention_weights = tf.nn.softmax(self.V(score), axis=1)
        
        print("Here 4")
        # context_vector shape after sum == (batch_size, hidden_size)
        context_vector = attention_weights * enc_output
        context_vector = tf.reduce_sum(context_vector, axis=1)
        #print("Hallo jeee")
        
        print("Here 5")
        # x shape after passing through embedding == (batch_size, 1, embedding_dim)
        x = self.bert_preprocess(x)
        print("Preproc op: ",x['input_word_ids'].shape)
        #print("Hallo Ahuhuh")
        print("Here 6")
        x = self.bert_model(x)["pooled_output"]
        print("Model op: ",x.shape)
        print("Here 7")
        #print("Ahuhuh")
        x = tf.expand_dims(x,axis=1)
        print("Model op expanded: ",x.shape)

        print("Here 8")
        # x shape after concatenation == (batch_size, 1, embedding_dim + hidden_size)
        x = tf.concat([tf.expand_dims(context_vector, 1), x], axis=-1)
        print("Model op concat: ",x.shape)

        print("Here 9")
        # passing the concatenated vector to the GRU
        output, state = self.gru(x)
        print("GRU op: ",output.shape)
        print("GRU state op: ",state.shape)
        
        print("Here 10")
        # output shape == (batch_size * max_length, hidden_size)
        output = tf.reshape(output, (-1, output.shape[2]))
        print("GRU op reshaped: ",output.shape)

        print("Here 11")
        # output shape == (batch_size * max_length, vocab)
        x = self.fc(output)
        print("Here 12")
        print("Final op: ",x.shape)
        
        return x, state, attention_weights
```

```python
optimizer = tf.keras.optimizers.Adam()
loss_object = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True, reduction='none')
def loss_function(real, pred,loss_object):

    mask = tf.math.logical_not(tf.math.equal(real, 0))
    loss_ = loss_object(real, pred)

    mask = tf.cast(mask, dtype=loss_.dtype)
    loss_ *= mask

    return tf.reduce_mean(loss_)
```
